### Festagent.com 


```python
begin = time.time()
webD = wb.Chrome(executable_path=r"C:\Users\Admin\OneDrive\Sagasit\chromedriver.exe") #web driver
listOflinks=[]
websiteLinks=[]
eve_dates=[]
fest_name=[]
yr=[]
sh_desc=[]
loc=[]
df = pd.DataFrame(columns = ['Festival Name','Years running','Event Dates','Short Description','Location','Festival URL','Festival website URL',])
for i in range(1,62):#looping through 61 pages
    #url=base_link+i
    #ids=[]
    webD.get("https://festagent.com/en/festivals?sort_by=name&page="+str(i)) #link for each page
    a=webD.find_elements_by_class_name('title-link') #elements of class 'title-link'
    b=webD.find_elements_by_class_name('festival-website') #Festival website link elements
    for el in a:
        tl=el.find_elements_by_tag_name('a')[0] #title link anchor tag
        listOflinks.append(tl.get_property('href')) #each festival link for more info
        tl1=tl.get_property('href')
        ss=tl1.split('festivals/')[-1]
        try:
            eve_dates.append(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[2]/p').text)
        except:
            #print("Incorrect path for "+ss+" event date")
            eve_dates.append('NA') 
        try:
            sh_desc.append(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[2]/div[2]/div[4]').text)
        except:
            sh_desc.append(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[2]/div[2]/div[3]').text)
            #print(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[2]/div[2]/div[3]').text)
        try:
            yr.append(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[2]/div[2]/div[1]').text)
        except:
            #print("Incorrect path for "+ss+" years running")
            yr.append('NA')
        try:
            loc.append(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[3]/p').text)
        except:
            #print("Incorrect path for "+ss+" location")
            loc.append('NA')
        try:
            fest_name.append(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[2]/div[2]/div[2]/a').text)
        except:
            fest_name.append(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[2]/div[2]/div[1]/a').text)
            #print(webD.find_element_by_xpath('//*[@id="'+ss+'"]/div[2]/div[2]/div[1]/a').text)
            #print("Incorrect path for "+ss+" festival name")
            #fest_name.append('NA')
    for elem in b:
        wl=elem.find_elements_by_tag_name('a')[0]
        websiteLinks.append(wl.get_property('href'))
    time.sleep(2)
df['Festival Name'] = fest_name
df['Years running'] = yr
df['Event Dates'] = eve_dates
df['Short Description'] = sh_desc
df['Location'] = loc
df['Festival URL'] = listOflinks
df['Festival website URL'] = websiteLinks
end = time.time()
print(end-begin)
df.to_csv('Festivals.csv',index=False)
```

### WHED.net

```python
begin=time.time()
#region variables
countries=[]
#countries=['Austria','Argentina','Canada (all)','China','India','Saudi Arabia','United Arab Emirates','United States of America (all)']
result = [] # List to store all data
univ_links=[] # Links for all universities
fields = ['Street:','City:','Province:','Post Code:','WWW:','Fields of study:','Job title:'] #Subfields list
#endregion variables

webD = wb.Chrome(executable_path=r'C:\Users\Admin\OneDrive\Sagasit\chromedriver.exe') # To launch chrome and run script
# Trigger the target website
webD.get("https://www.whed.net/results_institutions.php")
webD.implicitly_wait(5)

#all_countries=[]
cntry_el = webD.find_elements_by_xpath('//*[@id="Chp1"]/option')
#cntry_grp = webD.find_elements_by_xpath('//*[@id="Chp1"]/optgroup')
grps=webD.find_elements_by_xpath('//*[@id="Chp1"]/optgroup/option[1]')
for c in cntry_el:countries.append(c.text)
for g in grps: countries.append(g.text)
    
for cntry in countries:
    select = Select(webD.find_element_by_id('Chp1'))#select country dropdown
    select.select_by_visible_text(cntry)#choosing country

    Btn_GO = webD.find_element_by_xpath('//*[@id="fsearch"]/p/input')
    Btn_GO.click()

    select_rpp = Select(webD.find_element_by_name('nbr_ref_pge'))#select results per page drop down
    select_rpp.select_by_visible_text('100')#choosing 100 results per page option

    university_form = webD.find_element_by_xpath('//*[@id="contenu"]').find_element_by_id('results') 
    university_list = university_form.find_elements_by_xpath('//*[@id="results"]/li') # list of university elements
    for univ in range(len(university_list)):
            href = university_list[univ].find_element_by_class_name('details').find_elements_by_tag_name('a')[0].get_property('href') # University details link
            univ_links.append(href)
    while True:
        try:
            webD.find_element_by_partial_link_text('Next').click()
            university_form = webD.find_element_by_xpath('//*[@id="contenu"]').find_element_by_id('results')
            university_list = university_form.find_elements_by_xpath('//*[@id="results"]/li')
            for univ in range(len(university_list)):
                href = university_list[univ].find_element_by_class_name('details').find_elements_by_tag_name('a')[0].get_property('href') # University details link
                univ_links.append(href)
        except NoSuchElementException: break
    
for l in univ_links:
    webD.get(l) # Goes to each university details page
    webD.implicitly_wait(2)
    title=webD.find_element_by_xpath('//*[@id="page"]/div/div/div[2]/div[1]').text
    title_detailed = webD.find_element_by_xpath('//*[@id="page"]/div/div/div[2]/div[2]').text
    cntry_name=webD.find_element_by_xpath('//*[@id="contenu"]/p[2]').text
    t1=webD.find_elements_by_class_name('dt') # Main categories
    t2=webD.find_elements_by_class_name('dd')        
    labels=webD.find_elements_by_class_name('libelle') # Sub categories 
    content=webD.find_elements_by_class_name('contenu') 
    temp={}
    fos=''
    fos1=''
    temp.update({'Title': title,'Detailed Title':title_detailed,'Country':cntry_name})
    # Main categories info
    for i in range(len(t1)):
        if t1[i].text == '' or t1[i].text == 'Address':
            continue 
        else:
            value=t2[i].text
            temp.update({t1[i].text:value.replace('\n',',')})
    #Sub categories info
    for j in range(len(content)):
        if labels[j].text in fields:
            if labels[j].text == 'Fields of study:':
                info=content[j].text
                fos=fos+','+info
            elif labels[j].text == 'Job title:':
                info1=content[j].text
                fos1=fos1+','+info1
                #print(fos1)
            else:
                key=labels[j].text
                temp.update({key[:-1]: content[j].text})
        
    temp.update({'Fields of study': fos.lstrip(','),'Job titles':fos1.lstrip(',')})
    result.append(temp)
data=pd.DataFrame(result)
data
end=time.time()
print("Time taken : "+ str(end-begin) +"s")
data.to_csv("WHED1.csv",index=False)
```
